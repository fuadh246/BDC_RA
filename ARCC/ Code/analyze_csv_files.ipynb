```python
import os
import pandas as pd
from prettytable import PrettyTable
```

```python
folder_path = '../csv_file/'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
```

```python
csv_analysis = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        df = pd.read_csv(csv_path)

        # Analyze the CSV file
        analysis = {
            'File Name': filename,
            'Headers': list(df.columns),
            'Row Count': df.shape[0],
            'Column Count': df.shape[1],
            'Summary Statistics': df.describe(include='all').to_dict()
        }

        csv_analysis.append(analysis)

# Display the analysis results for each CSV file
for analysis in csv_analysis:
    print("File:", analysis['File Name'])
    print("Headers:", analysis['Headers'])
    print("Row Count:", analysis['Row Count'])
    print("Column Count:", analysis['Column Count'])
    print("Summary Statistics:")
    for column, stats in analysis['Summary Statistics'].items():
        print(f"  Column: {column}")
        for stat, value in stats.items():
            print(f"    {stat}: {value}")
    print("="*50)
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Define the folder path where CSV files are located

# Initialize a PrettyTable for display
table = PrettyTable()
table.field_names = ["File Name", "Headers",
                     "Row Count", "Column Count", "Summary Statistics"]

# Lists to store data for graphs
csv_filenames = []
row_counts = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        df = pd.read_csv(csv_path)

        # Analyze the CSV file
        analysis = {
            'File Name': filename,
            'Headers': ', '.join(list(df.columns)),
            'Row Count': df.shape[0],
            'Column Count': df.shape[1],
            'Summary Statistics': ', '.join([f"{col}: {desc}" for col, desc in df.describe(include='all').loc['mean'].items()])
        }

        table.add_row([analysis['File Name'], analysis['Headers'], analysis['Row Count'],
                       analysis['Column Count'], analysis['Summary Statistics']])

        # Add data for the graph
        csv_filenames.append(filename)
        row_counts.append(analysis['Row Count'])

# Print the PrettyTable
print(table)

# Choose graph type

print("1. Bar Graph")


# Create the chosen graph
plt.figure(figsize=(10, 6))
plt.bar(csv_filenames, row_counts, color='skyblue')
plt.xlabel('CSV File')
plt.ylabel('Row Count')
plt.title('Row Counts for Each CSV File')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("2. Pie Chart")

plt.figure(figsize=(8, 8))
plt.pie(row_counts, labels=csv_filenames,
        autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Distribution of Row Counts in CSV Files')
plt.show()
```

```python
# Initialize a list to store the names of CSV files with row count less than 100
small_row_files = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        df = pd.read_csv(csv_path)

        # Check if the row count is less than 100
        if df.shape[0] < 100:
            small_row_files.append(filename)

# Display the names of CSV files with row count less than 100
if small_row_files:
    print("CSV files with row count less than 100:")
    for filename in small_row_files:
        print(filename)
else:
    print("No CSV files found with row count less than 100.")
```
