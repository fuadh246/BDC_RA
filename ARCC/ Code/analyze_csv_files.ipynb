```python
import os
import pandas as pd
from prettytable import PrettyTable
```

```python
folder_path = '../csv_file/'
csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]
```

```python
csv_analysis = []

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        df = pd.read_csv(csv_path)

        # Analyze the CSV file
        analysis = {
            'File Name': filename,
            'Headers': list(df.columns),
            'Row Count': df.shape[0],
            'Column Count': df.shape[1],
            'Summary Statistics': df.describe(include='all').to_dict()
        }

        csv_analysis.append(analysis)

# Display the analysis results for each CSV file
for analysis in csv_analysis:
    print("File:", analysis['File Name'])
    print("Headers:", analysis['Headers'])
    print("Row Count:", analysis['Row Count'])
    print("Column Count:", analysis['Column Count'])
    print("Summary Statistics:")
    for column, stats in analysis['Summary Statistics'].items():
        print(f"  Column: {column}")
        for stat, value in stats.items():
            print(f"    {stat}: {value}")
    print("="*50)
```

```python
# Initialize a PrettyTable for display
table = PrettyTable()
table.field_names = ["File Name", "Headers",
                     "Row Count", "Column Count", "Summary Statistics"]

# Loop through all CSV files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".csv"):
        csv_path = os.path.join(folder_path, filename)
        df = pd.read_csv(csv_path)

        # Analyze the CSV file
        analysis = {
            'File Name': filename,
            'Headers': ', '.join(list(df.columns)),
            'Row Count': df.shape[0],
            'Column Count': df.shape[1],
            'Summary Statistics': ', '.join([f"{col}: {desc}" for col, desc in df.describe(include='all').loc['mean'].items()])
        }

        table.add_row([analysis['File Name'], analysis['Headers'], analysis['Row Count'],
                       analysis['Column Count'], analysis['Summary Statistics']])

# Print the PrettyTable
print(table)
```
